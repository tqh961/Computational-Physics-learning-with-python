# Simple graph in python with detail explanation

```python
import numpy as np    # Import numpy
from pylab import *   # Import pylab
```

## Simple way to plot a graph

```python
# Setup the x and y data
x = np.array([1,2,3,4,5])
y = x + 3

# Plot y versus x, using red X symbol, rx means red x
plot(x, y, 'rx')

# Write title for the graph
title('First Python Graph')

# Label x-axis
xlabel('Time')

# Label y-axis
ylabel('Distance')

# Limit x-axis from 0 to 6
xlim(0,6)

# Limit y-axis from 0 to 10
ylim(0,10)
```

## Much complex way to plot a graph

```python
time = linspace(0.0, 10.0, 10000)        #Linear space for time from 0.0 through 10.0 with 100 of them
height = exp(-time/3.0)*sin(time*3)    #This line make an array called height with each element calculated from time
figure()                               #This command make a figure window
plot(time, height, 'm-^')              #Plot time and height. m-^ is color and symbol indicators
plot(time, 0.3*sin(time*3), 'g-')      #Second plot with height replace with formula
legend(['damped', 'constant amplitude'], loc = 'upper right') #Make legend on the upper right with loc
xlabel('Time (s)')                     #label x
```
